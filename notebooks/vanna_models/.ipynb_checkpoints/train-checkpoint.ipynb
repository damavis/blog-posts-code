```python
def clear_all_training_data():
    [vn.remove_training_data(id) for id in list(vn.get_training_data()['id'])]
```

```python
#Remove training data
clear_all_training_data()
```

```python
vn.get_training_data()
```

```python
vn.train(ddl="""
CREATE TABLE customers (
    customer_id INT PRIMARY KEY,
    name TEXT NOT NULL,
    email TEXT,
    phone TEXT,
    address TEXT,
    created_at TIMESTAMP DEFAULT NOW()
);
""")
```

```python
vn.train(ddl="""
CREATE TABLE orders (
    order_id INT PRIMARY KEY,
    customer_id INT,
    order_date TIMESTAMP DEFAULT NOW(),
    status TEXT,
    total_amount DECIMAL(10, 2),
    FOREIGN KEY (customer_id) REFERENCES customers(customer_id)
);
""")
```

```python
### Trainig DATA
```

```python
vn.train(documentation="Our business use the table customers to save our customers data and table orders to store our business customers orders, cannot exist an order without a customer reference")
```

```python
vn.train(sql="SELECT * FROM customers WHERE name = 'John Doe'")
```

```python
vn.train(
    question="¿Cuántos clientes tenemos registrados?",
    sql="SELECT COUNT(*) FROM customers"
)
```

```python
vn.train(
    question="¿Cuál es el monto total de todas las órdenes?",
    sql="SELECT SUM(total_amount) FROM orders"
)
```

```python
vn.train(
    question="¿Cuáles son los clientes que realizaron pedidos en una fecha específica?",
    sql="SELECT customers.* FROM customers JOIN orders ON customers.customer_id = orders.customer_id WHERE order_date::date = ?"
)
```

```python
vn.train(
    question="¿Cuáles son los clientes sin órdenes registradas?",
    sql="SELECT * FROM customers WHERE customer_id NOT IN (SELECT customer_id FROM orders)"
)
```
